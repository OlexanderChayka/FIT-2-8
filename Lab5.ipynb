```python
#Побудова графіка функції  

import math 

import numpy as np  

import matplotlib.pyplot as plt  

from scipy import optimize  

from scipy.misc import derivative  

# область значень для x та y  

x_min, x_max = -2, 4  
y_min, y_max = -2, 3  
step = 0.01  
# створюємо масиви значень x та y  
x, y = np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))  

# рівняння системи  
eq1 = np.sin(x+1) - y -1.2  
eq2 = 2*x + np.cos(y) -2  

fig, ax = plt.subplots(figsize=(10, 10)) # створюємо графік  
ax.contour(x, y, eq1, levels=[0], colors='brown') #графік першого рівняння  
ax.contour(x, y, eq2, levels=[0], colors='green') # графік другого рівняння  

# налаштування графіка  

ax.set_xlim([x_min, x_max])  
ax.set_ylim([y_min, y_max])  

#підписи осей 
ax.set_xlabel('x')  
ax.set_ylabel('y')  

ax.set_title('Графік системи рівнянь') #назва графіка 

plt.grid(True)  
```
